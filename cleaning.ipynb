```python
# Filter data for Region 1
dftemp = df.loc[df['Region'] == 1]

# Set figure size and style
plt.figure(figsize=(13, 6))
sns.set_style('whitegrid')

# Updated countplot with better colors
sns.countplot(x='month', hue='Classes', data=dftemp, palette='husl')  # Use 'husl' palette for distinct colors

# Titles and Labels
plt.ylabel('Number of Fires', weight='bold', fontsize=12)
plt.xlabel('Months', weight='bold', fontsize=12)
plt.title("Fire Analysis of Sidi-Bel Regions", weight='bold', fontsize=15)

# Adjust Legend Position
plt.legend(title='Classes', loc='upper right', fontsize=10)  # Move legend to avoid overlap

# Add gridlines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# Filter data for Region 1
dftemp = df.loc[df['Region'] == 0]

# Set figure size and style
plt.figure(figsize=(13, 6))
sns.set_style('whitegrid')

# Updated countplot with better colors
sns.countplot(x='month', hue='Classes', data=dftemp, palette='husl')  # Use 'husl' palette for distinct colors

# Titles and Labels
plt.ylabel('Number of Fires', weight='bold', fontsize=12)
plt.xlabel('Months', weight='bold', fontsize=12)
plt.title("Fire Analysis of Bijaia Regions", weight='bold', fontsize=15)

# Adjust Legend Position
plt.legend(title='Classes', loc='upper right', fontsize=10)  # Move legend to avoid overlap

# Add gridlines for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
